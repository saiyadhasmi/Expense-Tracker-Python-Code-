```python
# Python Mini Project
# Title: Personal Expense Tracker
# Objective: The objective of this project is to create a simple Python program that helps users track their daily expenses using variables, user input, and basic calculations.
#Problem Statement: Design a Python program that asks the user for their daily budget and expenses in different categories, then calculates the total amount spent and displays the remaining balance in a clear summary format.
#Program Requirements:The program should
# 1. Ask the user for: Name, Daily Budget, Food Expense, Transport Expense, Other Expense
# 2. Store all inputs in variables
# 3. Calculate: Total Expenses, Remaining Balance
#Display a formatted daily expense report
```

```python
# -------- Personal Expense Tracker (Improved Beginner Version) --------

# ================== Person 1 ==================
name_1 = "Hasmi"
print("\nName:", name_1)

daily_budget_1 = int(input("Enter daily budget: "))

# Food expenses
food_item1_1 = input("Enter food item 1 name: ")
food_price1_1 = int(input("Enter food item 1 price: "))

food_item2_1 = input("Enter food item 2 name: ")
food_price2_1 = int(input("Enter food item 2 price: "))

food_item3_1 = input("Enter food item 3 name: ")
food_price3_1 = int(input("Enter food item 3 price: "))

total_food_1 = food_price1_1 + food_price2_1 + food_price3_1

# Other expenses
transport_1 = int(input("Enter transport expense: "))
other_1 = int(input("Enter other expense: "))

total_spent_1 = total_food_1 + transport_1 + other_1
remaining_1 = daily_budget_1 - total_spent_1

print("\n------ Daily Expense Report ------")
print("Name:", name_1)
print("Budget:", daily_budget_1)
print("Total Food:", total_food_1)
print("Transport:", transport_1)
print("Other:", other_1)
print("Total Spent:", total_spent_1)
print("Remaining Balance:", remaining_1)
print("---------------------------------")


# ================== Person 2 ==================
name_2 = "Sahal"
print("\nName:", name_2)

daily_budget_2 = int(input("Enter daily budget: "))

food_item1_2 = input("Enter food item 1 name: ")
food_price1_2 = int(input("Enter food item 1 price: "))

food_item2_2 = input("Enter food item 2 name: ")
food_price2_2 = int(input("Enter food item 2 price: "))

food_item3_2 = input("Enter food item 3 name: ")
food_price3_2 = int(input("Enter food item 3 price: "))

total_food_2 = food_price1_2 + food_price2_2 + food_price3_2

transport_2 = int(input("Enter transport expense: "))
other_2 = int(input("Enter other expense: "))

total_spent_2 = total_food_2 + transport_2 + other_2
remaining_2 = daily_budget_2 - total_spent_2

print("\n------ Daily Expense Report ------")
print("Name:", name_2)
print("Budget:", daily_budget_2)
print("Total Food:", total_food_2)
print("Transport:", transport_2)
print("Other:", other_2)
print("Total Spent:", total_spent_2)
print("Remaining Balance:", remaining_2)
print("---------------------------------")


# ================== Person 3 ==================
name_3 = "Moin"
print("\nName:", name_3)

daily_budget_3 = int(input("Enter daily budget: "))

food_item1_3 = input("Enter food item 1 name: ")
food_price1_3 = int(input("Enter food item 1 price: "))

food_item2_3 = input("Enter food item 2 name: ")
food_price2_3 = int(input("Enter food item 2 price: "))

food_item3_3 = input("Enter food item 3 name: ")
food_price3_3 = int(input("Enter food item 3 price: "))

total_food_3 = food_price1_3 + food_price2_3 + food_price3_3

transport_3 = int(input("Enter transport expense: "))
other_3 = int(input("Enter other expense: "))

total_spent_3 = total_food_3 + transport_3 + other_3
remaining_3 = daily_budget_3 - total_spent_3

print("\n------ Daily Expense Report ------")
print("Name:", name_3)
print("Budget:", daily_budget_3)
print("Total Food:", total_food_3)
print("Transport:", transport_3)
print("Other:", other_3)
print("Total Spent:", total_spent_3)
print("Remaining Balance:", remaining_3)
print("---------------------------------")
```
